```python
aggregated_df = df.groupby('Crop')['Yield (Quintal/ Hectare) '].sum().reset_index()

# Convert the aggregated DataFrame to lists
crop_names = aggregated_df['Crop'].tolist()
total_yields = aggregated_df['Yield (Quintal/ Hectare) '].tolist()

# Print the lists
print("Crop Names List:")
print(crop_names)
print("Total Yields List:")
print(total_yields)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(crop_names, total_yields, color='skyblue')

# Customize the plot
plt.title('Total Yield by Crop')
plt.xlabel('Crop Name')
plt.ylabel('Total Yield')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
state_df = df.groupby('State')['Yield (Quintal/ Hectare) '].sum().reset_index()

# Convert the aggregated DataFrame to lists
state_names = state_df['State'].tolist()
total_yields = state_df['Yield (Quintal/ Hectare) '].tolist()

# Print the lists
print("Crop Names List:")
print(crop_names)
print("Total Yields List:")
print(total_yields)
```

```python
import pandas as pd

# Load the three CSV files

# Reading the data into pandas dataframes
datafile_3 = pd.read_csv("datafile (3).csv")
datafile = pd.read_csv("datafile.csv")
produce = pd.read_csv("produce.csv")

# Checking the first few rows of each file to understand their contents
datafile_3_head = datafile_3.head()
datafile_head = datafile.head()
produce_head = produce.head()

datafile_3_head, datafile_head, produce_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the visual styles
sns.set(style="whitegrid")

# Cleaning and selecting relevant data
datafile_clean = datafile.set_index('Crop').dropna()  # Drop rows with NaN and set 'Crop' as index
produce_clean = produce.dropna(subset=['Particulars'])  # Remove rows with missing 'Particulars'

# Visualizing crop production over time for selected crops (2004-2012)
years = ['2004-05', '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11', '2011-12']

plt.figure(figsize=(12, 8))
for crop in datafile_clean.index[:5]:
    plt.plot(years, datafile_clean.loc[crop, years], marker='o', label=crop)

plt.title('Crop Production Over Time (2004-2012)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Production (in % of baseline)')
plt.xticks(rotation=45)
plt.legend(title="Crop")
plt.tight_layout()
plt.show()

# Showing the total production by crop type from produce.csv for recent years
recent_years = ['3-2010', '3-2011', '3-2012', '3-2013', '3-2014']
produce_clean_recent = produce_clean[['Particulars'] + recent_years].dropna()

# Plotting total agricultural production for food grains
plt.figure(figsize=(10, 6))
sns.barplot(data=produce_clean_recent, x='Particulars', y='3-2014', palette="viridis")
plt.title('Agricultural Production by Crop Type (2014)', fontsize=14)
plt.ylabel('Production (Ton mn)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python

# Set up the visual styles
sns.set(style="whitegrid")

# Cleaning and selecting relevant data from datafile.csv
datafile_clean = datafile.set_index('Crop').dropna()  # Drop rows with NaN and set 'Crop' as index

# Visualizing crop production over time for selected crops (2004-2012)
years = ['2004-05', '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11', '2011-12']

plt.figure(figsize=(12, 8))
for crop in datafile_clean.index[:5]:  # Adjust the number of crops displayed as necessary
    plt.plot(years, datafile_clean.loc[crop, years], marker='o', label=crop)

plt.title('Crop Production Over Time (2004-2012)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Production (in % of baseline)')
plt.xticks(rotation=45)
plt.legend(title="Crop")
plt.tight_layout()
plt.show()

# Cleaning produce.csv and handling columns dynamically
produce_clean = produce.dropna(subset=['Particulars'])  # Remove rows with missing 'Particulars'
recent_years = ['3-2010', '3-2011', '3-2012', '3-2013', '3-2014']

# Checking for available columns in the dataset
available_years = [col for col in recent_years if col in produce_clean.columns]
produce_clean_recent = produce_clean[['Particulars'] + available_years].dropna()

# Plotting total agricultural production for food grains (recent years)
plt.figure(figsize=(10, 6))
sns.barplot(data=produce_clean_recent, x='Particulars', y='3-2014', palette="viridis")  # Change the year based on availability
plt.title('Agricultural Production by Crop Type (2014)', fontsize=14)
plt.ylabel('Production (Ton mn)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
