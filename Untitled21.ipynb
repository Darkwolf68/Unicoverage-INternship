```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("datafile (1).csv")
df1=pd.read_csv("datafile (2).csv")
df2=pd.read_csv("datafile (3).csv")
df3=pd.read_csv("datafile.csv")
df4=pd.read_csv("produce.csv")
```

```python
df.info()
df1.info()
df2.info()
df3.info()
df4.info()
```

```python
df.info()
```

```python
df.rename(columns={'Cost of Cultivation (`/Hectare) A2+FL':'Cost of Cultivation per Heactare'},inplace=True)
```

```python
df['Total cost of cultivation']=df['Cost of Cultivation per Heactare']+ df['Cost of Cultivation (`/Hectare) C2']
```

```python
df.info()
```

```python
print(df.columns)
```

```python
total_cost = df['Total cost of cultivation']
total_yield = df['Yield (Quintal/ Hectare) ']


plt.figure(figsize=(8, 6))
plt.scatter(total_cost, total_yield, color='b', alpha=0.7)

# Customize the plot
plt.title('Yield (Quintal/ Hectare)  vs. Total cost of cultivation')
plt.xlabel('Total cost of cultivation')
plt.ylabel('Yield (Quintal/ Hectare) ')
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
total_cost = df['Cost of Production (`/Quintal) C2']
total_yield = df['Yield (Quintal/ Hectare) ']


plt.figure(figsize=(8, 6))
plt.scatter(total_cost, total_yield, color='b', alpha=0.7)

# Customize the plot
plt.title('Yield (Quintal/ Hectare)  vs. Cost of Production (`/Quintal) C2')
plt.xlabel('Total cost of production')
plt.ylabel('Yield (Quintal/ Hectare) ')
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
print(df['Crop'].unique())
```

```python
aggregated_df = df.groupby('Crop')['Yield (Quintal/ Hectare) '].sum().reset_index()

# Convert the aggregated DataFrame to lists
crop_names = aggregated_df['Crop'].tolist()
total_yields = aggregated_df['Yield (Quintal/ Hectare) '].tolist()

# Print the lists
print("Crop Names List:")
print(crop_names)
print("Total Yields List:")
print(total_yields)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(crop_names, total_yields, color='skyblue')

# Customize the plot
plt.title('Total Yield by Crop')
plt.xlabel('Crop Name')
plt.ylabel('Total Yield')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
state_df = df.groupby('State')['Yield (Quintal/ Hectare) '].sum().reset_index()

# Convert the aggregated DataFrame to lists
state_names = state_df['State'].tolist()
total_yields = state_df['Yield (Quintal/ Hectare) '].tolist()

# Print the lists
print("Crop Names List:")
print(crop_names)
print("Total Yields List:")
print(total_yields)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(state_names, total_yields, color='skyblue')

# Customize the plot
plt.title('Total Yield by State')
plt.xlabel('State Name')
plt.ylabel('Total Yield')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
